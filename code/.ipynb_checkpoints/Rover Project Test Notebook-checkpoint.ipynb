```python
# Define a function to perform a perspective transform
# I've used the example grid image above to choose source points for the
# grid cell in front of the rover (each grid cell is 1 square meter in the sim)
def perspect_transform(img):
    
    img_size = (img.shape[1], img.shape[0])
    # Define calibration box in source (actual) and destination (desired) coordinates
    # These source and destination points are defined to warp the image
    # to a grid where each 10x10 pixel square represents 1 square meter
    dst_size = 5 
    # Set a bottom offset to account for the fact that the bottom of the image 
    # is not the position of the rover but a bit in front of it
    bottom_offset = 6
    src = np.float32([[14, 140], [301 ,140],[200, 96], [118, 96]])
    dst = np.float32([[img_size[0]/2 - dst_size, img_size[1] - bottom_offset],
                      [img_size[0]/2 + dst_size, img_size[1] - bottom_offset],
                      [img_size[0]/2 + dst_size, img_size[1] - 2*dst_size - bottom_offset], 
                      [img_size[0]/2 - dst_size, img_size[1] - 2*dst_size - bottom_offset],
                      ])
       
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, img_size)# keep same size as input image
    mask = cv2.warpPerspective(np.ones_like(img[:,:,0]), M, img_size)
    return warped, mask

warped, mask = perspect_transform(image)
plt.imshow(warped)
scipy.misc.imsave('../../../classroom_visual_assets/warp_example_nogrid.jpg', warped)
```

```python
# Identify pixels below the threshold
# Threshold of RGB > 160 does a nice job of identifying ground pixels only
def rgb_thresh(img, rgb_thresh=(160, 160, 160)):    
    goodpix = (img[:,:,0] > rgb_thresh[0]) \
                & (img[:,:,1] > rgb_thresh[1]) \
                & (img[:,:,2] > rgb_thresh[2])
    color_select = np.zeros_like(img[:,:,0])
    color_select[goodpix] = 1
    return color_select


threshed = rgb_thresh(warped)
xpix, ypix, dist, angle = get_dist_angle(threshed, mask)
fig = plt.figure(figsize=(12,3))
plt.subplot(121)
plt.imshow(threshed, cmap='gray')
plt.subplot(122)
plt.plot(xpix, ypix, '.', markersize=3)
plt.xlim(-160, 160)
plt.ylim(0, 160)
#plt.imshow(threshed, cmap='gray')
#scipy.misc.imsave('../../../classroom_visual_assets/warped_threshed.jpg', threshed*255)
```

```python
# Define a different color selection scheme to find the rock samples
# All the rocks are the same color and the following color selection
# Works well to identify only rock samples
def find_rocks(img, levels=(110, 110, 50)):
    rockpix = ((img[:,:,0] > levels[0]) \
                & (img[:,:,1] > levels[1]) \
                & (img[:,:,2] < levels[2]))
    
    color_select = np.zeros_like(img[:,:,0])
    color_select[rockpix] = 1
    hotrocks = rockpix.nonzero()
    
    # If there's a rock in the frame, locate it in rover-centric coords
    if hotrocks[0].any():
        # Calculate rock pixel x and y coords in rover-centric reference frame
        rocky = np.absolute(hotrocks[0] - (img.shape[0] - 1)).astype(np.float)
        rockx = (hotrocks[1] - (img.shape[0] - 1)).astype(np.float)
        # Find the pixel position of the bottom of the rock
        bottom = np.argmin(rocky)
        rockymin = rocky[bottom]
        rockxmin = rockx[bottom]
        # Calculate distance and angle to rock
        rockdist = np.sqrt(rockxmin**2 + rockymin**2)
        rockang = np.arctan2(rockxmin, rockymin)
        rock_angdist = (rockang, rockdist)
    else:
        rock_angdist = None
        
    return rock_angdist, color_select

rock_angdist, rock_binary = find_rocks(rock_img, levels=(110, 110, 50))
fig = plt.figure(figsize=(12,3))
plt.subplot(121)
plt.imshow(rock_img)
plt.subplot(122)
plt.imshow(rock_binary, cmap='gray')
```

```python
# Run this cell repeatedly to try all steps on a random series of images.

# Define a function to calculate distances and angles to all "on" pixels 
# in a binary image assuming the camera center is at the midpoint in x and max in y
# This is assuming the image passed in is the thresholded warped image
def get_dist_angle(binary, mask, onpix=1):
    if onpix == 1:
        xy = binary.nonzero()
    else:
        xy = ((binary == 0) & (mask == 1)).nonzero()
    # Calculate pixel positions with reference to the rover position being at the 
    # center bottom of the image.  
    ypix = np.absolute(xy[0]-(binary.shape[0] - 1)).astype(np.float)
    xpix = (xy[1] - (binary.shape[0] - 1)).astype(np.float)
    # Calculate distance to each pixel
    dist = np.sqrt(xpix**2 + ypix**2)
    # Calculate angle away from vertical for each pixel
    angle = np.arctan2(xpix, ypix)
    return xpix, ypix, dist, angle

# Define a function to take a histogram of angular values
def angular_hist(angles, nbins=16):
    angle_hist = np.histogram(angles, bins=nbins)
    hist_values = angle_hist[0]
    binsize = angle_hist[1][1] - angle_hist[1][0]
    bin_centers = angle_hist[1][:-1] + binsize/2
    peak_idx = np.argmax(hist_values)
    peak_dir = bin_centers[peak_idx]
    peak_val = hist_values[peak_idx]
    min_idx = np.argmax(hist_values)
    min_dir = bin_centers[min_idx]
    min_val = hist_values[min_idx]
    total_nav = np.sum(hist_values)
    mean_dir = np.sum(hist_values*bin_centers)/np.sum(hist_values)
    steer_angle = (2*mean_dir + peak_dir)/3/(np.pi/2)
    return peak_dir, peak_val, min_dir, min_val, mean_dir, steer_angle, hist_values, bin_centers, binsize


# Grab another random image
#idx = np.random.randint(0, len(img_list)-1)
#image = mpimg.imread(img_list[idx])
warped, mask = perspect_transform(image)
threshed = rgb_thresh(warped)
rock_angdist, rock_img = find_rocks(warped, levels=(110, 110, 50))

# Calculate pixel values in rover-centric coords and distance/angle to all pixels
xpix, ypix, dist, angle = get_dist_angle(threshed, mask)
# Calculate angular histogram and outputs for deciding throttle, brake and steering
peak_dir, peak_val, min_dir, min_val, mean_dir, steer_angle, hist_values, bin_centers, binsize = angular_hist(angle)

fig = plt.figure(figsize=(12,9))
plt.subplot(321)
plt.imshow(image)
plt.subplot(322)
plt.imshow(warped)
plt.subplot(323)
plt.imshow(threshed, cmap='gray')
plt.subplot(324)
plt.imshow(rock_img, cmap='gray')
plt.subplot(325)
plt.plot(xpix, ypix, '.')

plt.xlim(-160, 160)
plt.ylim(0, 160)
max_angle = angle[np.argmax(hist_values)]
# Define an arrow that's 50 pixels long
arrow_length = 100
x_arrow = arrow_length * np.sin(max_angle)
y_arrow = arrow_length * np.cos(max_angle)
plt.arrow(0, 0, x_arrow, y_arrow, color='red', zorder=2, head_width=10, width=2)
print(x_arrow, y_arrow)
plt.subplot(326)
#plt.plot(angle, dist, '.', markersize=2)
plt.bar(bin_centers, hist_values, binsize)
```

```python
# Define a function to pass stored images to
# This function will be used by moviepy to create an output video
def process_image(img):
    
    warped, mask = perspect_transform(img)
    threshed = rgb_thresh(warped, rgb_thresh=(160, 160, 160))
    rock_location, rock_img = find_rocks(warped, levels=(110, 110, 50))
    
    if data.count < len(data.images):
        # Get pixel values and angle/distance measurements to navigable surface pixels
        x_nav, y_nav, dist_nav, ang_nav = get_dist_angle(threshed, mask)
        # Get pixel values and angle/distance measurements to obstacle pixels
        x_obs, y_obs, dist_obs, ang_obs = get_dist_angle(threshed, mask, onpix=0)
        
        # Compute angular histogram values for navigable pixels data
        peak_dir, peak_val, min_dir, min_val, mean_dir, steer_angle, hist_values, bin_centers = angular_hist(ang_nav)
        
        # Convert yaw values to +/- 180
        if data.yaw[data.count] <= 180:
            yaw_rover = data.yaw[data.count] * np.pi/180
        else: 
            yaw_rover = (data.yaw[data.count] - 360) * np.pi/180

        # Compute yaw values and x,y positions of navigable pixels in world space
        nav_angles = ang_nav + yaw_rover
        x_world = np.clip(np.int_((dist_nav * np.sin(nav_angles)) + (data.xpos[data.count]*10)), 0, 1999)
        y_world = np.clip(np.int_((dist_nav * np.cos(nav_angles)) + (data.ypos[data.count]*10)), 0, 1999)
        # Update the map
        data.worldmap[x_world, y_world, 2] += 10
        
        # Compute yaw values and x,y positions of navigable pixels in world space
        obs_angles = ang_obs + yaw_rover
        x_world = np.clip(np.int_((dist_obs * np.sin(obs_angles)) + (data.xpos[data.count]*10)), 0, 1999)
        y_world = np.clip(np.int_((dist_obs * np.cos(obs_angles)) + (data.ypos[data.count]*10)), 0, 1999)
        # Update the map
        data.worldmap[x_world, y_world, 0] += 0
        
        # Zero out obstacle pixels where votes for "navigable" at that location strongly outweigh votes for "obstacle"
        navigable = data.worldmap[:,:,2]
        obstacle = data.worldmap[:,:,0]
        likely_nav = navigable >= obstacle
        obstacle[likely_nav] = 0
        data.worldmap[:, :, 0] = obstacle
        
        # Add rock position to map if any rocks were found in the frame
        if rock_location is not None:
            rock_angle = rock_location[0] + yaw_rover
            x_world = np.clip(np.int((rock_location[1] * np.sin(rock_angle)) + (data.xpos[data.count]*10)), 0, 1999)
            y_world = np.clip(np.int((rock_location[1] * np.cos(rock_angle)) + (data.ypos[data.count]*10)), 0, 1999)
            data.worldmap[(x_world-10):(x_world+10), (y_world-10):(y_world+10), 1] += 100
        
        # Update count to keep track of where were are in the data
        data.count += 1
    
    # Create output to show original image, processing and mapmaking
    out = np.zeros((img.shape[0]*2, img.shape[1]*2, 3))
    out[0:img.shape[0], 0:img.shape[1]] = img
    thresh_3d = np.dstack((threshed*255, threshed*255, threshed*255))

    # Create smaller scaled versions of navigable, obstacle and sample maps
    worldmap_small = cv2.resize(data.worldmap.clip(0, 255), (img.shape[0], img.shape[0]))
    out[0:img.shape[0], img.shape[1]:] = warped
    out[img.shape[0]:, img.shape[1]:] = thresh_3d
    out[img.shape[0]:, :img.shape[0]] = np.rot90(worldmap_small)
    
    cv2.putText(out,"Blue = navigable terrain", (img.shape[0]-20,img.shape[0]+20), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)
    cv2.putText(out,"Red = obstacles", (img.shape[0]-20,img.shape[0]+35), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)
    cv2.putText(out,"Whitish = sample positions", (img.shape[0]-20,img.shape[0]+50), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)

    return out
```

```python
# Define a function to pass stored images to
# This function will be used by moviepy to create an output video
def process_image2(img):
    
    warped, mask = perspect_transform(img)
    threshed = rgb_thresh(warped, rgb_thresh=(160, 160, 160))
    rock_location, rock_img = find_rocks(warped, levels=(110, 110, 50))
    mini_thresh = cv2.resize(threshed, (32, 16))
    mini_mask = cv2.resize(mask, (32, 16))
    mapsize = data.smallmap.shape[0]
    peak_val = 0
    steer_angle = 0
    
    if data.count < len(data.images):
        # Get pixel values and angle/distance measurements to navigable surface pixels
        x_nav, y_nav, dist_nav, ang_nav = get_dist_angle(mini_thresh, mini_mask)
        # Get pixel values and angle/distance measurements to obstacle pixels
        x_obs, y_obs, dist_obs, ang_obs = get_dist_angle(mini_thresh, mini_mask, onpix=0)
        
        # Compute angular histogram values for navigable pixels data
        peak_dir, peak_val, min_dir, min_val, mean_dir, steer_angle, hist_values, bin_centers = angular_hist(ang_nav, nbins=8)
        
        # Convert yaw values to +/- 180
        if data.yaw[data.count] <= 180:
            yaw_rover = data.yaw[data.count] * np.pi/180
        else: 
            yaw_rover = (data.yaw[data.count] - 360) * np.pi/180

        # Compute yaw values and x,y positions of navigable pixels in world space
        
        nav_angles = ang_nav + yaw_rover
        x_world = np.clip(np.int_((dist_nav * np.sin(nav_angles)) + (data.xpos[data.count])), 0, mapsize-1)
        y_world = np.clip(np.int_((dist_nav * np.cos(nav_angles)) + (data.ypos[data.count])), 0, mapsize-1)
        # Update the map
        data.smallmap[x_world, y_world, 2] += 10
        
        # Compute yaw values and x,y positions of navigable pixels in world space
        obs_angles = ang_obs + yaw_rover
        x_world = np.clip(np.int_((dist_obs * np.sin(obs_angles)) + (data.xpos[data.count])), 0, mapsize-1)
        y_world = np.clip(np.int_((dist_obs * np.cos(obs_angles)) + (data.ypos[data.count])), 0, mapsize-1)
        # Update the map
        data.smallmap[x_world, y_world, 0] += 10
        
        # Zero out obstacle pixels where votes for "navigable" at that location strongly outweigh votes for "obstacle"
        navigable = data.smallmap[:,:,2]
        obstacle = data.smallmap[:,:,0]
        likely_nav = ((navigable > 0) & (obstacle > 0)).nonzero()
        obstacle[likely_nav] = 0
        data.smallmap[:, :, 0] = obstacle
        
        # Add rock position to map if any rocks were found in the frame
        if rock_location is not None:
            rock_angle = rock_location[0] + yaw_rover
            rock_dist = rock_location[1]/10
            x_world = np.clip(np.int((rock_dist * np.sin(rock_angle)) + (data.xpos[data.count])), 0, mapsize-1)
            y_world = np.clip(np.int((rock_dist * np.cos(rock_angle)) + (data.ypos[data.count])), 0, mapsize-1)
            data.smallmap[(x_world-1):(x_world+1), (y_world-1):(y_world+1), 1] += 100
        
        # Update count to keep track of where were are in the data
        data.count += 1
    
    # Create output to show original image, processing and mapmaking
    out = np.zeros((mapsize * 2, img.shape[1] * 2, 3))
    out[0:img.shape[0], 0:img.shape[1]] = img
    thresh_3d = np.dstack((threshed*255, threshed*255, threshed*255))

    # Create smaller scaled versions of navigable, obstacle and sample maps
    #worldmap_small = cv2.resize(data.worldmap.clip(0, 255), (img.shape[0], img.shape[0]))
    out[0:img.shape[0], img.shape[1]:2*img.shape[1]] = warped
    out[img.shape[0]:2*img.shape[0], img.shape[1]:2*img.shape[1]] = thresh_3d
    out[img.shape[0]: img.shape[0] + mapsize, :mapsize] = np.rot90(np.clip(data.smallmap, 0, 255))
    
    cv2.putText(out,"Blue = navigable terrain", (img.shape[0]-20,img.shape[0]+20), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)
    cv2.putText(out,"Red = obstacles", (img.shape[0]-20,img.shape[0]+35), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)
    cv2.putText(out,"Whitish = sample positions", (img.shape[0]-20,img.shape[0]+50), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)
    cv2.putText(out,"Histogram Max:"+str(peak_val), (img.shape[0]+30,img.shape[1]), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)
    cv2.putText(out,"Steer Angle:"+str(steer_angle), (img.shape[0]+30,img.shape[1]+15), 
                cv2.FONT_HERSHEY_COMPLEX, 0.4, (255, 255, 255), 1)

    return out
```
